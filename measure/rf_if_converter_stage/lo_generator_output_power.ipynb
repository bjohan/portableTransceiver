```python
import sys
sys.path.append("gpib_instrument_control")
sys.path.append("lmx2594/py")

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]

import skrf as rf
import scipy.io as sio
import gpib_instrument_control.wavetek_8502
import gpib_instrument_control.hp_8700_series_vna
import lmx2594cmd2
import time 

#pll = lmx2594cmd2.Lmx2594('/dev/ttyUSB2', fosc=280e6, mash_order=3)
#vna = gpib_instrument_control.hp_8700_series_vna.Hp8753A()
ps = gpib_instrument_control.wavetek_8502.Wavetek8502()
```

```python
nMHz = 20e3-750;
freqs = np.linspace(750e6, 20e9, int(nMHz/10)+1);
amps = []
vna.setSParameter('S11')
for f in freqs:
    vna.cw(f)
    time.sleep(0.5)
    amps.append(ps.readChannel(f))
plt.plot(freqs, amps)
plt.grid(True)
plt.xlabel('Frequency Hz')
plt.ylabel('Output power [dBm]')
sio.savemat('vna_port1_0dBm_power.mat', {'frequencies':freqs, 'output_power':amps})
```

```python
import numpy as np
pll = lmx2594cmd2.Lmx2594('/dev/ttyUSB0', fosc=280e6, mash_order=3)
nMHz = 15e3-750;
freqs = np.linspace(750e6, 15e9, int(nMHz/10)+1);
amps = []

for f in freqs:
    pll.setCwFreq(f)
    amps.append(ps.readChannel(f))
    
plt.plot(freqs, amps)
plt.grid(True)
plt.xlabel('Frequency Hz')
plt.ylabel('Output power [dBm]')

sio.savemat('lo_2nd_high_power_a_pos.mat', {'frequencies':freqs, 'output_power':amps})
```
